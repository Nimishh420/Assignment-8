```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("wine.csv")
df.head(10)
```

```python
#Wine types values count
df['Type'].value_counts()
```

```python
#Dropping Wine Type column 
df2=df.iloc[:,1:]
df2.head(10)
```

```python
df2.shape
```

```python
df2.info()
```

```python
# Converting data to numpy array
df_ary=df2.values
df_ary
```

```python
# Normalizing the numerical data 
from sklearn.preprocessing import scale
df_norm=scale(df_ary)
df_norm
```

```python
# Applying PCA Fit Transform to dataset
from sklearn.decomposition import PCA
pca=PCA(n_components=13)

df_pca=pca.fit_transform(df_norm)
df_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color='k')
```

```python
# Final Dataframe
final_df=pd.concat([df['Type'],pd.DataFrame(df_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

**Checking with other Clustering Algorithms**

**1. Hierarchical Clustering**

```python
# Import Libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# create Dendrograms
plt.figure(figsize=(30,15))
dendrogram=sch.dendrogram(sch.linkage(df_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters




```

```python
y=pd.DataFrame(hclusters.fit_predict(df_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
df3=df.copy()
df3['clustersid']=hclusters.labels_
df3
```

**2. K-Means Clustering**

```python
from sklearn.cluster import KMeans
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm  K=3
clusters3=KMeans(3,random_state=30).fit(df_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
df4=df.copy()
df4['clusters3id']=clusters3.labels_
df4
```
